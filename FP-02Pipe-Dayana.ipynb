### Reading data

```python
path ='data\Train.csv'
test_path='data\Test.csv'
df= pd.read_csv(path, index_col='Tour_ID')
X_val=pd.read_csv(test_path, index_col='Tour_ID')
df.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['cost_category'], axis=1), df['cost_category'], 
                                                    test_size=0.1,random_state=0, shuffle=True, 
                                                    stratify= df['cost_category'])
```

### Defining steps for the final pipeline

#### preprocessing steps

```python
from sklearn.preprocessing import FunctionTransformer, OneHotEncoder, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

# Early preprocessing -> using business Logic
#fillAlone transformer definition
def early_proc(df):
    df['travel_with'] = df.apply(lambda row: 'Alone'
                                 if (pd.isnull(row['travel_with']) & (row['total_male']+row['total_female']==1))
           else row['travel_with'], axis=1)
    df.drop(['tour_arrangement'], axis=1, inplace=True)
    return df

early_proc_tr= FunctionTransformer(early_proc)
# fill_alone_pipe = Pipeline('fill_alone_tr', fill_alone_tr)


# country pipe definition
def clean_country(df):
    df['country'].replace({'BURGARIA':'BULGARIA', 'DRC':'CONGO', 'UAE':'UNITED ARAB EMIRATES'}, inplace= True)
    return df
clean_country_tr= FunctionTransformer(clean_country)

country_pipe = Pipeline(steps=[('clean_country_tr', clean_country_tr),
                                  ('ohe', OneHotEncoder(handle_unknown='ignore', sparse= False))
                                  ])


#travel with pipeline
travel_with_pipe = Pipeline(steps=[
    ('most_imp', SimpleImputer(strategy='most_frequent')),
    ('ohe', OneHotEncoder(handle_unknown='ignore',sparse=False))
                                  ])

##activity pipe
def activity_pipe(df):
    df['main_activity'].replace({'Widlife Tourism':'Wildlife Tourism'}, inplace= True)
    return df
activity_pipe_tr= FunctionTransformer(activity_pipe)

activity_pipe = Pipeline(steps=[('activity_pipe_tr', activity_pipe_tr),
                                  ('ohe', OneHotEncoder(handle_unknown='ignore', sparse= False))
                                  ])

##numerical pipe
num_var = ['total_female','total_male','night_mainland','night_zanzibar']
num_pipe = SimpleImputer(strategy='mean')

##
cat_var=['age_group','purpose', 'info_source']
cat_pipe=OneHotEncoder(handle_unknown='ignore', sparse=False)

##
bin_var=['package_transport_int','package_accomodation','package_food','package_transport_tz','package_sightseeing',
         'package_guided_tour','package_insurance', 'first_trip_tz']
bin_pipe=OneHotEncoder(sparse=False, drop='first')


preproc_ct=ColumnTransformer(transformers=[
#     ('fillAlone', fill_alone_tr, ['travel_with', 'total_male', 'total_female'])
    ('travelWith_pipe', travel_with_pipe, ['travel_with']),
    ('country_pipe', country_pipe, ['country']),
    ('activity_pipe', activity_pipe, ['main_activity']),
    ('num_pipe', num_pipe, num_var),
    ('cat_pipe', cat_pipe, cat_var),
    ('bin_var', bin_pipe, bin_var)],
                             remainder='passthrough')
```

```python
from sklearn.metrics import classification_report
import xgboost as xgb
```

#### scaling step definition

```python
#scaling step definition

from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler, LabelEncoder

min_max_scaler=MinMaxScaler()
```

### Putting steps together

```python
#final pipeline definition
pipe=Pipeline(steps=[
    ('fillAlone_n_drop', early_proc_tr), #early preprocessing step
    ('clean_n_enc', preproc_ct), #preprocessing step
    ('scal', min_max_scaler), #scaling step
    ('model', xgb.XGBClassifier()) #modeling step
])
```

```python
#encodig target variable
targ_enc=LabelEncoder()
y_train_enc=targ_enc.fit_transform(y_train)
y_test_enc=targ_enc.transform(y_test)
```

```python
#fitting pipeline
pipe.fit(X_train, y_train_enc)
```

```python
# saving model
import joblib
joblib.dump(pipe, 'Dayana_model-1.joblib')
```

```python
#getting prediction on test set
y_pred_test_enc=pipe.predict(X_test)
```

```python
#calculate score on test set
print("TEST classification report:\n",classification_report(y_test_enc,y_pred_test_enc))
```

```python
y_pred_val_enc=pipe.predict(X_val)
```

```python
#reverse encoding
y_pred_val=targ_enc.inverse_transform(y_pred_val_enc)
y_pred_val
```

```python
submission_df=pd.DataFrame(y_pred_val, index=X_val.index, columns=['PredictedClass'])
display(pd.DataFrame(submission_df))
```
