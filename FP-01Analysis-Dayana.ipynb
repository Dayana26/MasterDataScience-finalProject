# Just Simple Exploratory ModelTrials
### Trials with no imputing

#### Simple preprocessing
1. fillAlone and dropNa
2. Predict with the best DecisionTree and then  RandomForest

```python
df['travel_with'] = df.apply(lambda row: 'Alone' if 
                             (pd.isnull(row['travel_with']) & (row['total_male']+row['total_female']==1))
                             else row['travel_with'], axis=1)
```

```python
drp_df=df.dropna().copy(deep=True)
drp_df.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(drp_df.drop(['cost_category'], axis=1), drp_df['cost_category'], 
                                                    test_size=0.1,random_state=0, shuffle=True, 
                                                    stratify= drp_df['cost_category'])
```

```python
#preprocessing data
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer

# ord_enc=OrdinalEncoder(dtype=int, handle_unknown='error'
# #                        , unknown_value= 6
#                       )
ord_col=['age_group'] #not took in consideration because its worse impacts on predict scoring

lab_enc=OneHotEncoder(drop='first', dtype=int)
lab_mask=['package_transport_int', 'package_accomodation', 'package_food','package_transport_tz', 
          'package_sightseeing', 'package_guided_tour', 'package_insurance', 'first_trip_tz']

oh_enc= OneHotEncoder() #could be improved for column 'travel_with'
cat_mask = X_train.drop( lab_mask, axis=1).select_dtypes(exclude='number').columns

enc_transf=ColumnTransformer([
#     ('ord_enc', ord_enc, ord_col),
    ('lab_enc', lab_enc, lab_mask ), ('oh_enc', oh_enc, cat_mask )],
                             remainder='passthrough')
X_train_t = enc_transf.fit_transform(X_train)
X_test_t=enc_transf.transform(X_test)
```

```python
X_test_t.shape
```

```python
columns=enc_transf.get_feature_names()
X_train_t_df= pd.DataFrame(data=X_train_t.toarray(), columns= columns)
X_train_t_df
```

### Modeling Trials

####  Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, f1_score


#Decision Tree classifier
dt_clf= DecisionTreeClassifier()
param={'criterion':['gini', 'entropy'],
       'min_samples_split': [2, 3, 5],
       'ccp_alpha': [0.0, 0.01]
      }
cv_split=StratifiedKFold(n_splits=6,random_state=42, shuffle=True )
grid_search= RandomizedSearchCV(estimator=dt_clf, param_distributions=param, cv=cv_split,n_iter=2)
dt_gs= grid_search.fit(X_train_t, y_train)
dt_model=dt_gs.best_estimator_
```

```python
print('Best Decision Tree esimator params: ')
display(dt_model.get_params())
```

```python
##printing results

print('DECISION TREE SCORING \n')
y_pred_dt=dt_model.predict(X_test_t)
print("TEST classification report:\n",classification_report(y_test,y_pred_dt))

y_pred_train_dt=dt_model.predict(X_train_t)
print("TRAIN classification report:\n",classification_report(y_train,y_pred_train_dt))
```

####  Random Forest Classifier

```python
%%time
#RandomForest Classifier

from sklearn.ensemble import RandomForestClassifier

rf_clf= RandomForestClassifier()
rf_param={'n_estimators': [150],
       'criterion': ['gini'],
#        'max_depth'=[],
    'min_samples_split': [2, 3, 5],
#     min_samples_leaf=1,
#     min_weight_fraction_leaf=0.0,
#     max_features='auto',
#     max_leaf_nodes=None,
#     min_impurity_decrease=0.0,
#     min_impurity_split=None,
#     bootstrap=True,
#     oob_score=False,
#     n_jobs=None,
#     random_state=None,
#     verbose=0,
#     warm_start=False,
#     class_weight=None,
    'ccp_alpha': [0.0, 0.01]
#     max_samples=None,
      }

rf_grid_search= RandomizedSearchCV(estimator=rf_clf, param_distributions=rf_param, cv=cv_split, n_iter=1)
rf_grid_search.fit(X_train_t, y_train)
rf_model=rf_grid_search.best_estimator_
```

```python
print('Best Random Forest esimator params: ')
display(rf_model.get_params())
```

```python
##printing results

print('RANDOM FOREST SCORING \n')

y_pred_rf=rf_model.predict(X_test_t)
print("TEST classification report:\n",classification_report(y_test,y_pred_rf))

y_pred_train_rf=rf_model.predict(X_train_t)
print("TRAIN classification report:\n",classification_report(y_train,y_pred_train_rf))
```
